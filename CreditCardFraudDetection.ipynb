SPLITTING INTO TRAINING AND TESTING

```python
from sklearn.model_selection import train_test_split
```

```python
x=data.drop(['Class'],axis=1)
```

```python
y=data['Class']
```

```python
print("Shape of x:", x.shape)
print("Shape of y:", y.shape)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=123)
```

```python
print("Shape of x_train:", x_train.shape)
print("Shape of y_train:", y_train.shape)
```

```python
print("Shape of x_test:", x_test.shape)
print("Shape of y_test:", y_test.shape)
```

# Model 1: Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
model1=lr.fit(x_train,y_train)
```

```python
prediction1=model1.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm1= confusion_matrix(y_test,prediction1)
```

```python
sns.heatmap(cm1,annot=True,fmt='g')
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
accuracy1=accuracy_score(y_test,prediction1)
print("Accuracy of Logistic Regression Model:",accuracy1)
```

```python
pre1=precision_score(y_test,prediction1)
print("Precision of Logistic Regression Model:",pre1)
```

```python
re1=recall_score(y_test,prediction1)
print("Recall of Logistic Regression Model:",re1)
```

```python
f1_1=f1_score(y_test,prediction1)
print("F1_score of Logistic Regression Model:",f1_1)
```

# Model 2: Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc= DecisionTreeClassifier()
```

```python
model2=dtc.fit(x_train,y_train)
```

```python
prediction2=model2.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm2= confusion_matrix(y_test,prediction2)
```

```python
sns.heatmap(cm2,annot=True,fmt='g')
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
accuracy2=accuracy_score(y_test,prediction2)
print("Accuracy of Decision Tree Model:",accuracy2)
```

```python
pre2=precision_score(y_test,prediction2)
print("Precision of Decision Tree Model:",pre2)
```

```python
re2=recall_score(y_test,prediction2)
print("Recall of Decision Tree Model:",re2)
```

```python
f1_2=f1_score(y_test,prediction2)
print("F1_score of Decision Tree Model:",f1_2)
```

# Exporting the model

```python
import pickle

# Save the Decision Tree model to a pickle file
with open('decision_tree_model.pkl', 'wb') as f:
    pickle.dump(model2, f)

print("Model saved successfully as 'decision_tree_model.pkl'")
```
